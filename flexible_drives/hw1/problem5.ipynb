Roller chain is to be used in a hydraulic forklift truck to elevate the forks. If two strands support the load equally which size would you specify for a design load of 2.5 Tonnes? What will be your final safety factor?


I will assume we are dealing with imperial _tons_ here.

Each strand will support a load of 1.25 tons as the load is split equally.

We allow each strand to use only 10% of their max tensile strength from table 7-12.



```python
load_per_strand = 1.25 * 2000 # lbf
print(load_per_strand)
```

```python
min_strength = load_per_strand * 10
print(min_strength)
```

Since we need at least 25000 lbf breaking force, the minimum supported chain is #120.

As a result, our safety factor is 13.6

```python
f_s = 34000 / load_per_strand
print(f_s)
```
